```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping
```

```python
import urllib.request
import os
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00228/smsspamcollection.zip"
filename = "smsspamcollection.zip"

if not os.path.exists(filename):
    urllib.request.urlretrieve(url, filename)
```

```python
import zipfile
with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall()
```

```python
df = pd.read_csv('SMSSpamCollection', sep='\t', header=None, names=['label', 'message'])

print("ğŸ“Š Ø¨ÙŠØ§Ù†Ø§Øª Dataset:")
print(f"Ø¹Ø¯Ø¯ Ø§Ù„Ø¹ÙŠÙ†Ø§Øª: {len(df)}")
print(f"ØªÙˆØ²ÙŠØ¹ Ø§Ù„ØªØµÙ†ÙŠÙØ§Øª:\n{df['label'].value_counts()}")
```

```python
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
```

```python
texts = df['message'].values
labels = df['label'].values

print(f"\nğŸ“ Ù†Ù…ÙˆØ°Ø¬ Ù…Ù† Ø§Ù„Ù†ØµÙˆØµ:")
for i in range(3):
    print(f"{labels[i]}: {texts[i][:100]}...")
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    texts, labels, test_size=0.2, random_state=42, stratify=labels
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

print(f"\nğŸ“Š ØªÙˆØ²ÙŠØ¹ Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª:")
print(f"Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ¯Ø±ÙŠØ¨: {len(X_train)}")
print(f"Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„ØªØ­Ù‚Ù‚: {len(X_val)}")
print(f"Ø§Ù„Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±: {len(X_test)}")
```

```python
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)

X_train_seq = tokenizer.texts_to_sequences(X_train)
X_val_seq = tokenizer.texts_to_sequences(X_val)
X_test_seq = tokenizer.texts_to_sequences(X_test)
```

```python
seq_lengths = [len(seq) for seq in X_train_seq]
max_length = int(np.percentile(seq_lengths, 95))

X_train_pad = pad_sequences(X_train_seq, maxlen=max_length, padding='post')
X_val_pad = pad_sequences(X_val_seq, maxlen=max_length, padding='post')
X_test_pad = pad_sequences(X_test_seq, maxlen=max_length, padding='post')

print(f"\nğŸ”¢ Ø¥Ø¹Ø¯Ø§Ø¯Ø§Øª Ø§Ù„Ù†Ù…ÙˆØ°Ø¬:")
print(f"Ø­Ø¬Ù… Ø§Ù„Ù…ÙØ±Ø¯Ø§Øª: {len(tokenizer.word_index)}")
print(f"Ø§Ù„Ø·ÙˆÙ„ Ø§Ù„Ø£Ù…Ø«Ù„ Ù„Ù„ØªØ³Ù„Ø³Ù„: {max_length}")
```

```python
vocab_size = len(tokenizer.word_index) + 1
embedding_dim = 100

model = Sequential([
    Embedding(vocab_size, embedding_dim, input_length=max_length),

    Conv1D(128, 5, activation='relu'),
    # Removed GlobalMaxPooling1D() here as it reduces dimensions prematurely

    Conv1D(64, 3, activation='relu'),
    GlobalMaxPooling1D(), # Apply GlobalMaxPooling1D after all Conv1D layers

    Dense(64, activation='relu'),
    Dropout(0.5),

    Dense(32, activation='relu'),
    Dropout(0.3),

    Dense(1, activation='sigmoid')
])
model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy', 'precision', 'recall']
)

print("\nğŸ§® Ù…Ù„Ø®Øµ Ø§Ù„Ù†Ù…ÙˆØ°Ø¬:")
model.summary()
```

```python
early_stopping = EarlyStopping(patience=5, restore_best_weights=True)

history = model.fit(
    X_train_pad, y_train,
    validation_data=(X_val_pad, y_val),
    epochs=50,
    batch_size=32,
    callbacks=[early_stopping],
    verbose=1
)
```

```python
print("\nğŸ“ˆ ØªÙ‚ÙŠÙŠÙ… Ø§Ù„Ù†Ù…ÙˆØ°Ø¬ Ø¹Ù„Ù‰ Ø¨ÙŠØ§Ù†Ø§Øª Ø§Ù„Ø§Ø®ØªØ¨Ø§Ø±:")
test_loss, test_accuracy, test_precision, test_recall = model.evaluate(X_test_pad, y_test, verbose=0)
```

```python
y_pred_proba = model.predict(X_test_pad)
y_pred = (y_pred_proba > 0.5).astype(int).flatten()
```

```python
from sklearn.metrics import f1_score, roc_auc_score

test_f1 = f1_score(y_test, y_pred)
test_auc = roc_auc_score(y_test, y_pred_proba)

print(f"\nğŸ¯ Ù†ØªØ§Ø¦Ø¬ Ø§Ù„Ø£Ø¯Ø§Ø¡:")
print(f"Ø§Ù„Ø¯Ù‚Ø© (Accuracy): {test_accuracy:.4f}")
print(f"Ø§Ù„Ø¯Ù‚Ø© (Precision): {test_precision:.4f}")
print(f"Ø§Ù„Ø§Ø³ØªØ¯Ø¹Ø§Ø¡ (Recall): {test_recall:.4f}")
print(f"F1-Score: {test_f1:.4f}")
print(f"AUC-ROC: {test_auc:.4f}")
```

```python
print("\nğŸ“Š ØªÙ‚Ø±ÙŠØ± Ø§Ù„ØªØµÙ†ÙŠÙ Ø§Ù„ØªÙØµÙŠÙ„ÙŠ:")
print(classification_report(y_test, y_pred, target_names=['Ham', 'Spam']))
```
